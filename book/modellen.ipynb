Stel dat we de positie van een wagen observeren over een periode $\Delta t$. De wagen is
hierbij het systeem en de gemeten posities (bv. X en Y coordinaten) zijn de data.
Als de wagen geparkeerd staat binnen de gemeten periode, is het systeem **stationair**
(niet dynamisch) en zullen alle waarden $\vec{p}_i = (p_{x_i}, p_{y_i})$ identiek (*constant*) zijn. Als het
voertuig zich echter verplaatst zullen we uiteraard veranderende coordinaten observeren.  
:::

In veel toepassingen willen we aan de hand van data een zo goed mogelijk zicht krijgen op de voorbije,
de huidige en vooral ook de toekomstige interne states van een systeem. Daarom gaan we op zoek naar een
**wiskundige beschrijving van de relatie tussen onze data en het systeem dat ze voortbracht**.  

:::{important} ðŸ’¡
:icon: false
:class: simple 
In het voorbeeld van een wagen die zich verplaatst binnen $\Delta t$ en we dus veranderlijke
coordinaten zullen observeren, wordt de exacte waarde van de data op tijdstip $t_i$ bepaald door:  
- de start positie $\vec{p}_0$
- de snelheid (in X en Y richting) bij de start van de metingen $\vec{v}_0$
- de voorafgaande sequentie van versnellingen (in X en Y richting) $\vec{a}_{0\rightarrow i-1}$  

De specifieke relatie tussen positie en snelheid is de wiskundige integratie:
$$
\vec{p}_{i+1} = \vec{p}_i + \vec{v}_i \times dt
$$
en tussen snelheid en acceleratie:
$$
\vec{v}_{i+1} = \vec{v}_i + \vec{a}_i \times dt
$$
waarbij we uitgaan van een constante *sample rate* $dt = t_{i+1} - t_i$
:::

Deze wiskundige beschrijvingen vormen het **model van het systeem**. Aan de hand van dit model kunnen
we, gegeven de data, specifieke inzichten verwerven over de interne states.  
  
:::{important} ðŸ’¡
:icon: false
:class: simple
In het voorbeeld wordt het model ons gegeven door mechanische wetten, namelijk [Newtoniaanse
wetten van beweging](https://en.wikipedia.org/wiki/Newton%27s_laws_of_motion).
:::

toekomstige states?

:::{important} ðŸ’¡

Vertrekkende van de positie data op de verschillende tijdstippen $\vec{p}_{0\rightarrow i-1}$,
kunnen we uit het model afleiden dat de start snelheid gelijk is aan:
$$
\vec{v}_0=\frac{\vec{p}_1 - \vec{p}_0}{dt}
$$
en dat de versnelling bij de twee positie gelijk is aan:
$$
\vec{a}_1=\frac{(\frac{\vec{p}_2 - \vec{p}_1}{dt}) - \vec{v}_0}{dt}
$$
door deze bewerkingen sequentieel uit te rekenen krijgen we niet alleen een zicht op alle states (positie, 
snelheid en acceleratie van de wagen), maar kunnen we uiteindelijk ook een accurate voorspelling maken waar de wagen zich
zal bevinden op het eerst volgende niet-geobserveerde tijdstip $t_{n+1}$.
:::  

In heel veel situaties is het verkrijgen van een model voor het systeem niet simpelweg een kwestie van
physische wetmatigheden. Denk bijvoorbeeld aan het systeem van een bedrijfsaandeel op de beurs. 
Als we door observaties van de koers een zicht willen krijgen op de interne state om van daaruit bijvoorbeeld
de toekomstige waardering te voorspellen, is het verre van voor de hand liggend om een bruikbare
wiskundige beschrijving te bekomen. Een ander voorbeeld ... computer vision?
